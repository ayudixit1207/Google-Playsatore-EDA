df['reviews'].str.isnumeric() aur df['reviews'].str.isnumeric().sum() me difference ye hai ki str.isnumeric() column ki har value ko individually check karta hai aur batata hai ki wo poori tarah numeric hai ya nahi, isliye iska output True/False ki series hoti hai. Jaise "123" ke liye True aur "abc" ke liye False. Wahin str.isnumeric().sum() un sab True/False values ko count kar deta hai, kyunki Python me True ko 1 aur False ko 0 maana jata hai. Isliye .sum() lagane ke baad hume total number milta hai ki column me kitni values numeric hain. Matlab pehla code har row ka result dikhata hai, jabki doosra code numeric rows ki total count batata hai.

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
## Convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

Is code ka kaam Installs aur Price columns ko clean karna hai. Sabse pehle chars_to_remove naam ki list me +, , aur $ jaise special characters rakhe gaye hain aur cols_to_clean list me un columns ke naam diye gaye hain jinhe clean karna hai. Phir nested for loop ki madad se har special character ko ek-ek karke dono columns me dhoonda jata hai aur str.replace(item, '') ke through usse remove kar diya jata hai. Iske baad 1,000+ 1000 ban jata hai aur $4.99 4.99 ban jata hai. Is tarah data sirf numeric values me convert ho jata hai, jisse baad me astype(int) ya astype(float) ka use karke data type badalna aur machine learning ya data analysis ke liye use karna aasaan ho jata hai.

```python
df_copy['Price'].unique()
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store

## What are the Top 5 most installed Apps in Each popular Categories ??

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.

```python
## Check outliers using IQR method for numerical columns
for col in ['Rating', 'Reviews', 'Size', 'Installs', 'Price']:
    Q1 = df_copy[col].quantile(0.25)
    Q3 = df_copy[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df_copy[(df_copy[col] < lower) | (df_copy[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
## boxplot to see outliers
plt.figure(figsize=(15,10))
for i, col in enumerate(['Rating','Reviews','Size','Installs','Price']):
    plt.subplot(2,3,i+1)
    sns.boxplot(y=df_copy[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

# Feature Engineering

```python
## label encoding - Free=0, Paid=1
df_copy['Type_encoded'] = df_copy['Type'].map({'Free':0, 'Paid':1})
df_copy['Type_encoded'].value_counts()
```

```python
## log transformation for skewed columns
## log1p handles 0 values safely
df_copy['Reviews_log'] = np.log1p(df_copy['Reviews'])
df_copy['Installs_log'] = np.log1p(df_copy['Installs'])
df_copy['Price_log'] = np.log1p(df_copy['Price'])
```

```python
## before vs after log transformation
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].hist(df_copy['Installs'], bins=50, color='salmon')
axes[0].set_title('Installs - Before')
axes[1].hist(df_copy['Installs_log'], bins=50, color='steelblue')
axes[1].set_title('Installs_log - After')
plt.tight_layout()
plt.show()
```

```python
## correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df_copy.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
